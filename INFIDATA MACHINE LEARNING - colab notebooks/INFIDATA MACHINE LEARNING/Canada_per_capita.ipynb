```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```
uploaded=files.upload()
```

```
import io
capita = pd.read_csv(io.BytesIO(uploaded['canada_per_capita_income.csv']))
```

```
capita
```

```
x = capita['year']
y = capita['per capita income (US$)']
```

```
plt.scatter(x,y,color="blue",marker='*')
```

```
capita_new = capita.drop('per capita income (US$)',axis='columns')
capita_new
```

```
income = capita['per capita income (US$)']
income
```

```
from sklearn import linear_model
```

```
reg = linear_model.LinearRegression()
reg.fit(capita_new,income)
reg.score(capita_new,income)
```

```
plt.xlabel('Years')
plt.ylabel('Income')
plt.scatter(x,y,color='blue',marker='*')
plt.plot(capita_new,reg.predict(capita_new),color='red')
```
