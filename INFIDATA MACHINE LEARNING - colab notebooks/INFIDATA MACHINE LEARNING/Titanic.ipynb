```
import pandas as pd
```

```
from google.colab import files
uploaded = files.upload()
```

```
import io
df = pd.read_csv(io.BytesIO(uploaded['titanic.csv']))
df
```

```
df.groupby('Survived').mean()
```

```
X = df[['Pclass','Sex','Age','Fare']]
y = df['Survived']
```

```
X
```

```
from sklearn.preprocessing import LabelEncoder
le_Pclass = LabelEncoder()
le_Sex = LabelEncoder()
le_Age = LabelEncoder()
le_Fare = LabelEncoder()
```

```
X['Pclass_n']=le_Pclass.fit_transform(X['Pclass'])
X['Sex_n']=le_Sex.fit_transform(X['Sex'])
X['Age_n']=le_Age.fit_transform(X['Age'])
X['Fare_n']=le_Fare.fit_transform(X['Fare'])
```

```
X
```

```
X.drop(['Pclass','Sex','Age','Fare'],axis='columns',inplace=True)
```

```
X
```

```
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```
model.fit(X,y)
```

```
model.score(X,y)
```
