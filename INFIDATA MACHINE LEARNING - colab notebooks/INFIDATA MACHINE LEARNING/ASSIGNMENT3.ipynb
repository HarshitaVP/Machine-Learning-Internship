```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
from google.colab import files
uploaded = files.upload()
```

```
import io
covid = pd.read_excel(io.BytesIO(uploaded['Covid.xlsx']))
```

```
barWidth = 0.25
fig = plt.subplots(figsize =(25, 8))
```

```
br1 = np.arange(len(covid['Confirmed'])) 
br2 = [x + barWidth for x in br1] 
br3 = [x + barWidth for x in br2] 
br4 = [x + barWidth for x in br3]

barWidth = 0.25
fig = plt.subplots(figsize =(25, 8))


# Make the plot 
plt.bar(br1, covid['Confirmed'], color ='r', width = barWidth, 
        edgecolor ='grey', label ='Confirmed') 
plt.bar(br2, covid['Active'], color ='g', width = barWidth, 
        edgecolor ='grey', label ='Active') 
plt.bar(br3, covid['Recovered'], color ='b', width = barWidth, 
        edgecolor ='grey', label ='Recovered')
plt.bar(br4, covid['Deceased'], color ='k', width = barWidth, 
        edgecolor ='grey', label ='Deceased')

plt.xlabel("states")
plt.ylabel("Number of people")

plt.legend( labels = ['Confirmed','Active','Recovered','Deceased'])
```
