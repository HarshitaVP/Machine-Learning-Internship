```
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```
from google.colab import files
uploaded = files.upload()
```

```
import io
df = pd.read_csv(io.BytesIO(uploaded['homeprices (1).csv']))
df.head()
```

```
df.price
```

```
type(df.price)
```

```
df[['area']]
```

```
type(df[['area']])
```

```
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```
model.coef_
```

```
model.intercept_
```

```
model.predict([[5000]])
```

```
import pickle
```

```
with open('model_pickle_aug_2020','wb') as file:
    pickle.dump(model,file)
```

```
with open('model_pickle_aug_2020','rb') as file:
    mp = pickle.load(file)
```

```
mp.coef_
```

```
mp.intercept_
```

```
mp.predict([[5000]])
```
